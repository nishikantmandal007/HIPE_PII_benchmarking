# HIPE-2022 Dataset Analysis

This notebook analyzes the processed HIPE-2022 dataset, providing metrics on file counts and named entity distributions across various dimensions:
- Document types (DOCS, SRT, LOG)
- Dataset aliases
- Languages
- Splits (dev, train, test)

The analysis includes both file counts and person name entity statistics.

```python
import os
import glob
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict, Counter

# Set the style for our plots
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = [12, 8]
```

## 1. Configure Data Paths

```python
# Set the base directory for the processed dataset
BASE_DIR = "/home/stark007/MayaDataPrivacy/TESTAPI/DatasetTransformation/output"

# Define the document types and their corresponding directories
DOC_TYPES = {
    "docs": os.path.join(BASE_DIR, "docs"),
    "docsjson": os.path.join(BASE_DIR, "docsjson"),
    "srt": os.path.join(BASE_DIR, "srt"),
    "srtjson": os.path.join(BASE_DIR, "srtjson"),
    "log": os.path.join(BASE_DIR, "log"),
    "logjson": os.path.join(BASE_DIR, "logjson")
}

# Check if directories exist
for doc_type, dir_path in DOC_TYPES.items():
    print(f"{doc_type}: {'Directory exists' if os.path.exists(dir_path) else 'Directory NOT found'}")
```

## 2. Count Files by Document Type, Alias, Language, and Split

```python
def count_files_by_structure(base_dir):
    """Count files organized in the directory structure: base_dir/alias/language/split/"""
    counts = defaultdict(lambda: defaultdict(lambda: defaultdict(int)))
    total_count = 0
    
    # Get all aliases
    aliases = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]
    
    for alias in aliases:
        alias_dir = os.path.join(base_dir, alias)
        languages = [d for d in os.listdir(alias_dir) if os.path.isdir(os.path.join(alias_dir, d))]
        
        for lang in languages:
            lang_dir = os.path.join(alias_dir, lang)
            splits = [d for d in os.listdir(lang_dir) if os.path.isdir(os.path.join(lang_dir, d))]
            
            for split in splits:
                split_dir = os.path.join(lang_dir, split)
                file_count = len(os.listdir(split_dir))
                counts[alias][lang][split] = file_count
                total_count += file_count
    
    return counts, total_count

# Collect file counts for each document type
doc_type_counts = {}
doc_type_totals = {}

for doc_type, dir_path in DOC_TYPES.items():
    if os.path.exists(dir_path):
        doc_type_counts[doc_type], doc_type_totals[doc_type] = count_files_by_structure(dir_path)
    else:
        doc_type_counts[doc_type] = {}
        doc_type_totals[doc_type] = 0

# Create a summary DataFrame for file counts
summary_data = []

for doc_type, counts in doc_type_counts.items():
    for alias, lang_counts in counts.items():
        for lang, split_counts in lang_counts.items():
            for split, count in split_counts.items():
                summary_data.append({
                    "Document Type": doc_type,
                    "Dataset Alias": alias,
                    "Language": lang,
                    "Split": split,
                    "File Count": count
                })

summary_df = pd.DataFrame(summary_data)

# Display the total counts by document type
print("Total file counts by document type:")
for doc_type, total in doc_type_totals.items():
    print(f"{doc_type}: {total} files")
print(f"Grand total: {sum(doc_type_totals.values())} files")
```

## 3. File Distribution Visualizations

```python
# Visualize the distribution of files across document types
plt.figure(figsize=(10, 6))
sns.barplot(x=list(doc_type_totals.keys()), y=list(doc_type_totals.values()))
plt.title('Total Files by Document Type')
plt.xlabel('Document Type')
plt.ylabel('File Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualize file distribution by dataset alias
alias_counts = summary_df.groupby('Dataset Alias')['File Count'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='Dataset Alias', y='File Count', data=alias_counts)
plt.title('Total Files by Dataset Alias')
plt.xlabel('Dataset Alias')
plt.ylabel('File Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualize file distribution by language
lang_counts = summary_df.groupby('Language')['File Count'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Language', y='File Count', data=lang_counts)
plt.title('Total Files by Language')
plt.xlabel('Language')
plt.ylabel('File Count')
plt.tight_layout()
plt.show()

# Visualize file distribution by split
split_counts = summary_df.groupby('Split')['File Count'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Split', y='File Count', data=split_counts)
plt.title('Total Files by Split')
plt.xlabel('Split')
plt.ylabel('File Count')
plt.tight_layout()
plt.show()
```

## 4. File Count Summary Tables

```python
# Create pivot tables to summarize the data

# Files by dataset alias and language
alias_lang_pivot = summary_df.pivot_table(
    index='Dataset Alias', 
    columns='Language', 
    values='File Count', 
    aggfunc='sum',
    fill_value=0
)
alias_lang_pivot['Total'] = alias_lang_pivot.sum(axis=1)
alias_lang_pivot.loc['Total'] = alias_lang_pivot.sum()

# Files by dataset alias and split
alias_split_pivot = summary_df.pivot_table(
    index='Dataset Alias', 
    columns='Split', 
    values='File Count', 
    aggfunc='sum',
    fill_value=0
)
alias_split_pivot['Total'] = alias_split_pivot.sum(axis=1)
alias_split_pivot.loc['Total'] = alias_split_pivot.sum()

# Files by language and split
lang_split_pivot = summary_df.pivot_table(
    index='Language', 
    columns='Split', 
    values='File Count', 
    aggfunc='sum',
    fill_value=0
)
lang_split_pivot['Total'] = lang_split_pivot.sum(axis=1)
lang_split_pivot.loc['Total'] = lang_split_pivot.sum()

# Files by document type and dataset alias
doctype_alias_pivot = summary_df.pivot_table(
    index='Document Type', 
    columns='Dataset Alias', 
    values='File Count', 
    aggfunc='sum',
    fill_value=0
)
doctype_alias_pivot['Total'] = doctype_alias_pivot.sum(axis=1)
doctype_alias_pivot.loc['Total'] = doctype_alias_pivot.sum()

print("Files by Dataset Alias and Language:")
display(alias_lang_pivot)

print("\nFiles by Dataset Alias and Split:")
display(alias_split_pivot)

print("\nFiles by Language and Split:")
display(lang_split_pivot)

print("\nFiles by Document Type and Dataset Alias:")
display(doctype_alias_pivot)
```

## 5. Named Entity Analysis

```python
def extract_entities_from_json(json_dir):
    """Extract entity counts from JSON files"""
    entity_data = []
    
    # Get all JSON files recursively
    json_files = glob.glob(os.path.join(json_dir, '**/*.json'), recursive=True)
    
    for json_file in json_files:
        # Extract path components
        rel_path = os.path.relpath(json_file, json_dir)
        path_parts = rel_path.split(os.sep)
        
        if len(path_parts) >= 4:
            alias = path_parts[0]
            language = path_parts[1]
            split = path_parts[2]
            filename = path_parts[3]
            
            try:
                with open(json_file, 'r', encoding='utf-8') as f:
                    json_data = json.load(f)
                
                # Extract entities based on the file structure
                entity_count = 0
                entities = set()
                
                if "entities" in json_data:
                    # Structure with an "entities" key (srtjson, logjson)
                    if isinstance(json_data["entities"], dict):
                        # For structure like {"entities": {"page_1": [...], "page_2": [...]}}
                        for page, page_entities in json_data["entities"].items():
                            entities.update(page_entities)
                    elif isinstance(json_data["entities"], list):
                        # For structure like {"entities": [...]}
                        entities.update(json_data["entities"])
                else:
                    # For structure like {"page_1": [...], "page_2": [...]}
                    for page in ["page_1", "page_2"]:
                        if page in json_data:
                            entities.update(json_data[page])
                
                entity_count = len(entities)
                
                entity_data.append({
                    "File": filename,
                    "Dataset Alias": alias,
                    "Language": language,
                    "Split": split,
                    "Entity Count": entity_count,
                    "Unique Entities": len(entities)
                })
            except Exception as e:
                print(f"Error processing {json_file}: {e}")
    
    return pd.DataFrame(entity_data)

# Extract entity data from each JSON directory
entity_dfs = {}
for doc_type in ["docsjson", "srtjson", "logjson"]:
    if os.path.exists(DOC_TYPES[doc_type]):
        print(f"Extracting entity data from {doc_type}...")
        entity_dfs[doc_type] = extract_entities_from_json(DOC_TYPES[doc_type])
        print(f"Found {len(entity_dfs[doc_type])} files with entity data")
    else:
        print(f"Directory for {doc_type} not found")

# Compare entity counts across JSON types
if len(entity_dfs) > 1:
    # Add document type column to each dataframe
    for doc_type, df in entity_dfs.items():
        df['Document Type'] = doc_type
    
    # Combine all dataframes
    all_entities_df = pd.concat(entity_dfs.values(), ignore_index=True)
else:
    # If only one type was processed, use that
    doc_type = list(entity_dfs.keys())[0]
    all_entities_df = entity_dfs[doc_type]
    all_entities_df['Document Type'] = doc_type
```

## 6. Entity Count Summary Tables

```python
# Create summary tables for entity counts

# Total entity counts by dataset alias
entity_by_alias = all_entities_df.groupby('Dataset Alias')['Entity Count'].agg(['sum', 'mean', 'count']).rename(
    columns={'sum': 'Total Entities', 'mean': 'Avg Entities per File', 'count': 'File Count'}
)
entity_by_alias['Total Entities'] = entity_by_alias['Total Entities'].astype(int)
entity_by_alias['Avg Entities per File'] = entity_by_alias['Avg Entities per File'].round(2)

# Total entity counts by language
entity_by_lang = all_entities_df.groupby('Language')['Entity Count'].agg(['sum', 'mean', 'count']).rename(
    columns={'sum': 'Total Entities', 'mean': 'Avg Entities per File', 'count': 'File Count'}
)
entity_by_lang['Total Entities'] = entity_by_lang['Total Entities'].astype(int)
entity_by_lang['Avg Entities per File'] = entity_by_lang['Avg Entities per File'].round(2)

# Total entity counts by split
entity_by_split = all_entities_df.groupby('Split')['Entity Count'].agg(['sum', 'mean', 'count']).rename(
    columns={'sum': 'Total Entities', 'mean': 'Avg Entities per File', 'count': 'File Count'}
)
entity_by_split['Total Entities'] = entity_by_split['Total Entities'].astype(int)
entity_by_split['Avg Entities per File'] = entity_by_split['Avg Entities per File'].round(2)

# Entity counts by dataset alias and language
entity_alias_lang = all_entities_df.pivot_table(
    index='Dataset Alias',
    columns='Language',
    values='Entity Count',
    aggfunc='sum',
    fill_value=0
)
entity_alias_lang['Total'] = entity_alias_lang.sum(axis=1)
entity_alias_lang.loc['Total'] = entity_alias_lang.sum()

# Entity counts by dataset alias and split
entity_alias_split = all_entities_df.pivot_table(
    index='Dataset Alias',
    columns='Split',
    values='Entity Count',
    aggfunc='sum',
    fill_value=0
)
entity_alias_split['Total'] = entity_alias_split.sum(axis=1)
entity_alias_split.loc['Total'] = entity_alias_split.sum()

# If we have multiple document types, compare them
if 'Document Type' in all_entities_df.columns:
    entity_by_doctype = all_entities_df.groupby('Document Type')['Entity Count'].agg(['sum', 'mean', 'count']).rename(
        columns={'sum': 'Total Entities', 'mean': 'Avg Entities per File', 'count': 'File Count'}
    )
    entity_by_doctype['Total Entities'] = entity_by_doctype['Total Entities'].astype(int)
    entity_by_doctype['Avg Entities per File'] = entity_by_doctype['Avg Entities per File'].round(2)
    
    print("Entity counts by document type:")
    display(entity_by_doctype)

print("\nEntity counts by dataset alias:")
display(entity_by_alias)

print("\nEntity counts by language:")
display(entity_by_lang)

print("\nEntity counts by split:")
display(entity_by_split)

print("\nEntity counts by dataset alias and language:")
display(entity_alias_lang)

print("\nEntity counts by dataset alias and split:")
display(entity_alias_split)
```

## 11. Cross-File Type Entity Consistency Analysis

```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict

# Define the base output directory where your processed data is stored
output_base_dir = "/home/stark007/MayaDataPrivacy/TESTAPI/DatasetTransformation/output"

# Function to calculate folder size
def get_folder_size(path):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            if not os.path.islink(fp):
                total_size += os.path.getsize(fp)
    return total_size # Size in bytes

# --- 1. Load Dataset Statistics ---
print("--- Loading Dataset Statistics ---")
stats_file = os.path.join(output_base_dir, "dataset_statistics.json")
if os.path.exists(stats_file):
    with open(stats_file, 'r', encoding='utf-8') as f:
        dataset_statistics = json.load(f)
    print("Loaded dataset_statistics.json")
else:
    dataset_statistics = {}
    print("Warning: dataset_statistics.json not found. Some visualizations may be incomplete.")

# --- 2. Collect Information from Generated JSON Files (for names) ---
print("\n--- Collecting Information from Generated JSON Files (for names) ---")
file_data = []
for root, dirs, files in os.walk(os.path.join(output_base_dir, "docsjson")):
    for file in files:
        if file.endswith(".json"):
            filepath = os.path.join(root, file)
            parts = root.split(os.sep)
            
            try:
                alias_idx = parts.index("docsjson") + 1
                alias = parts[alias_idx]
                lang = parts[alias_idx + 1]
                split = parts[alias_idx + 2]
            except (ValueError, IndexError):
                alias = "unknown_alias"
                lang = "unknown_lang"
                split = "unknown_split"
            
            with open(filepath, 'r', encoding='utf-8') as f:
                data = json.load(f)
                
            page1_names = data.get("page_1", [])
            page2_names = data.get("page_2", [])
            
            file_data.append({
                "filename": file,
                "dataset_alias": alias,
                "language": lang,
                "split": split,
                "num_names_page1": len(page1_names),
                "num_names_page2": len(page2_names),
                "total_names": len(page1_names) + len(page2_names)
            })

files_df = pd.DataFrame(file_data)
print(f"Collected data for {len(files_df)} JSON files.")

# --- 3. Collect Information about all generated files for size and type breakdown ---
print("\n--- Collecting Information about all generated files ---")
all_generated_files = defaultdict(lambda: defaultdict(int)) # {file_type: {alias: count}}\nfile_sizes = defaultdict(int) # {file_type: total_size}\n
for root, dirs, files in os.walk(output_base_dir):
    if "visualizations" in root: # Skip visualizations directory
        continue
    for file in files:
        filepath = os.path.join(root, file)
        file_ext = os.path.splitext(file)[1].lstrip('.')
        file_type = file_ext if file_ext else "no_extension"
        
        # Try to extract alias from path for file type breakdown
        parts = root.split(os.sep)
        alias = "overall" # Default for overall count
        for part in parts:
            if part in dataset_statistics.get('alias_counts', {}): # Check if part is a known alias
                alias = part
                break
        
        all_generated_files[file_type][alias] += 1
        file_sizes[file_type] += os.path.getsize(filepath)

print("Collected file type and size information.")
```
